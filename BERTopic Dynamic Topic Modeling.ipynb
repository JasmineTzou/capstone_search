```python
!pip install pandas
import pandas as pd

!pip install nltk

import nltk
from nltk.corpus import stopwords
stopwords = list(set(stopwords.words('english')))
```

```python
text_list_sample = data
text_list_sample = pd.DataFrame({'text': text_list_sample})

text_list_sample['text']= text_list_sample['text'].apply(lambda x: x.lower())
```

```python

#stopwords = nltk.corpus.stopwords.words('english')
#stopwords =[ 'stop', 'the', 'to', 'and', 'a', 'in', 'it', 'is', 'I', 'that', 'had', 'on', 'for', 'were', 'was']

#defining the function to remove stopwords from tokenized text
def remove_stopwords(text):
  text = text.split()
  output= [i for i in text if i not in stopwords]
  return ' '.join(output)
#applying the function
text_list_sample['text']= text_list_sample['text'].apply(lambda x:remove_stopwords(x))

#importing the Stemming function from nltk library
from nltk.stem.porter import PorterStemmer
#defining the object for stemming
porter_stemmer = PorterStemmer()
#defining a function for stemming
def stemming(text):
  text = text.split()
  stem_text = [porter_stemmer.stem(word) for word in text]
  return ' '.join(stem_text)
text_list_sample['text']=text_list_sample['text'].apply(lambda x: stemming(x))
```

```python
labels = ["test", "said", "health", "lab", "covid19", "coronaviru","cnn","need", "worker", "medic"]
id2label = {}
label2id = {}
for label_id,label in enumerate(labels):
    id2label[label_id] = label
    label2id[label] = label_id
```

```python
from transformers import AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained('distilbert-base-uncased', use_fast=True)
```

```python
import datasets
my_dict = {"text": data}
dataset = datasets.Dataset.from_dict(my_dict)


tokenized_datasets = dataset.map(lambda x: tokenizer(x['text'], truncation=True), batched=True,remove_columns=['text'])
```

```python
from transformers import DataCollatorWithPadding

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

```python
from torch.utils.data import DataLoader


dataloader = DataLoader(
        dataset = tokenized_datasets,
        batch_size = 1,
        shuffle = True,
        collate_fn = data_collator,
    )
```

```python
from transformers import AutoModelForSequenceClassification


model = AutoModelForSequenceClassification.from_pretrained('distilbert-base-uncased', problem_type="multi_label_classification", num_labels=10, id2label=id2label, label2id=label2id)
```

```python
import torch
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
device
model.to(device)
```

```python
model.eval()
# results = []
# texts = []
# pred = []
# labels = []
with torch.no_grad():
  for batch in dataloader:
    # print(batch)
    # label = batch['labels'][0]
    batch = {k:v.to(device) for k,v in batch.items() if k in ['input_ids', 'attention_mask']}
    # print(batch['input_ids'][0])
    print(tokenizer.decode(batch['input_ids'][0]))
    outputs = model(**batch)
    predictions = torch.argmax(outputs.logits, dim=-1)
    # results.append(predictions.item() == label.item())
    # pred.append(id2label[predictions.item()])
    # labels.append(id2label[label.item()])
    # texts.append(tokenizer.decode(batch['input_ids'][0]))
    print('prediction: ' +  id2label[predictions.item()])
    # print('label: ' + id2label[label.item()])
```
